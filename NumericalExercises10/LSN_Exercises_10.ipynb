# <span style="color:darkred"> SOLUTION </span>

I solved option 1, namely the GA search with continents.

I performed the GA search in three conditions:
1. Single core optimization
2. Parallel independent searches (8 cores)
3. Parallel searches with migrations (8 cores)

In all cases I run the simulations for 1000 generations.
In order to compare the single core case with the others, I chose to use 250 individuals per core. Therefore in cases 2. and 3. there were 2000 individuals overall.

Other parameters of the problem are (same as LSN Exercise #9): probabilities of inversion, translation and switching 0.033 each; probability of crossover 0.80; cost function: $L^1$ norm.

For random number generation I used the parallel PRNG from the course's website.

The structure of the code is mostly derived from what I did in LSN Exercise #9.
There I used a PRNG from a C/C++ library, but now I reverted to the course's default PRNG. I extended it with a method for the uniform generation of integers on a given range, since I needed it.

The main difference is of course the implementation of the parallel searches with the MPI library. The code is structured as follows:

1. Initialization of the PRNG and loading of input data;
2. Random initialization of the populations for the parallel processes;
3. Gathering of the data from the continents to root and ranking of the global population;
4. Evolutionary algorithm; if GA search with migrations, every $N_\mathrm{migr}=10$ evolutionary steps a migration is performed.

The results are presented in the figures below. It can be seen that in the allocated generations the methods 1. and 2. do not converge to a path with no intersections, while the method with migrations does. I must emphasise that the number of individuals is here 250, while in LSN Exercise #9 I used 1000 individuals, which produced a loop free solution in the allocated generations. Here I wanted to emphasize the differences of the various approaches, and I therefore set out smaller populations for the GA searches.

The method with migrations is quite clearly the fastest converging one, in terms of number of generations needed.

```python
fig, axs = plt.subplots(2,2, figsize=(12,12), dpi=300)
x, y = np.loadtxt("./sol/ISOLATED_CONTINENTS/data/pos_US_caps.dat", usecols=(0,1), unpack=True)

axs[0,0].plot(x,y, alpha=0.5)
axs[0,0].scatter(x,y, s=10)
axs[0,0].set_title("Starting random solution - US state capitals")
axs[0,0].set_xlabel("Longitude")
axs[0,0].set_ylabel("Latitude")

x, y = np.loadtxt("./1core/data/route_out.dat", usecols=(0,1), unpack=True)
axs[0,1].plot(x,y, alpha=0.5)
axs[0,1].scatter(x,y, s=10)
axs[0,1].set_title("Optimized solution - 1 core")
axs[0,1].set_xlabel("Longitude")
axs[0,1].set_ylabel("Latitude")

x, y = np.loadtxt("./sol/ISOLATED_CONTINENTS/data/route_out.dat", usecols=(0,1), unpack=True)
axs[1,0].plot(x,y, alpha=0.5)
axs[1,0].scatter(x,y, s=10)
axs[1,0].set_title("Opt. solution - 8 cores - Isolated continents")
axs[1,0].set_xlabel("Longitude")
axs[1,0].set_ylabel("Latitude")

x, y = np.loadtxt("./sol/MIGRATION_CONTINENTS/data/route_out.dat", usecols=(0,1), unpack=True)
axs[1,1].plot(x,y, alpha=0.5)
axs[1,1].scatter(x,y, s=10)
axs[1,1].set_title("Opt. solution - 8 cores - Continents w/ migrations")
axs[1,1].set_xlabel("Longitude")
axs[1,1].set_ylabel("Latitude")

plt.show()
```
